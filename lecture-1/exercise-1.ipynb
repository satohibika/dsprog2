```python
import requests
from bs4 import BeautifulSoup
import sqlite3
import time

url = "https://github.com/psf/requests"
time.sleep(1) 
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

repo_name = soup.find("strong", {"itemprop": "name"}).text.strip() if soup.find("strong", {"itemprop": "name"}) else "不明"

language_tag = soup.select_one("li.d-inline:nth-child(1) > a > span")
language = language_tag.text.strip() if language_tag else "不明"

stars_tag = soup.select_one('a[href*="/stargazers"]')
stars = stars_tag.text.strip().replace(" stars", "") if stars_tag else "不明"

print("リポジトリ名:", repo_name)
print("言語:", language)
print("スター数:", stars)
```

```python
conn = sqlite3.connect("test.db")
cur = conn.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS repos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    language TEXT,
    stars TEXT
)
""")

cur.execute("DELETE FROM repos")

cur.execute("INSERT INTO repos (name, language, stars) VALUES (?, ?, ?)",
            (repo_name, language, stars))

conn.commit()
conn.close()
```

```python
conn = sqlite3.connect("test.db")
cur = conn.cursor()

cur.execute("SELECT * FROM repos")
rows = cur.fetchall()

for row in rows:
    print(row)

conn.close()
```
